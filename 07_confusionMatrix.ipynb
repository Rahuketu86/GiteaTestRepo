```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Example Data
total_samples = 150  # Total number of event days
TP = 50  # True Positives
FP = 10  # False Positives
FN = 20  # False Negatives
TN = 70  # True Negatives

# Derived Metrics
precision = TP / (TP + FP)
recall = TP / (TP + FN)
accuracy = (TP + TN) / total_samples

# Plotting Function
def plot_overlapping_squares(tp, fp, fn, tn, total):
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Outer square representing total event days
    env_square = patches.Rectangle((0, 0), 1, 1, edgecolor='black', facecolor='lightgrey', label="Total Samples")
    ax.add_patch(env_square)
    ax.text(0.5, 1.05, f"Total: {total}", ha='center', fontsize=12)
    
    # Square for actual positives
    actual_square = patches.Rectangle((0.2, 0.2), 0.6, 0.6, edgecolor='blue', facecolor='none', linewidth=2, label="Actual Positives")
    ax.add_patch(actual_square)
    ax.text(0.5, 0.85, f"Actual Positives: {tp + fn}", ha='center', fontsize=10, color='blue')
    
    # Square for predicted positives
    predicted_square = patches.Rectangle((0.35, 0.35), 0.4, 0.4, edgecolor='green', facecolor='none', linewidth=2, label="Predicted Positives")
    ax.add_patch(predicted_square)
    ax.text(0.65, 0.45, f"Predicted Positives: {tp + fp}", ha='center', fontsize=10, color='green')
    
    # TP Region (overlap)
    tp_square = patches.Rectangle((0.35, 0.35), 0.25, 0.25, edgecolor='red', facecolor='red', alpha=0.5, label="True Positives (TP)")
    ax.add_patch(tp_square)
    ax.text(0.48, 0.48, f"TP: {tp}", ha='center', fontsize=10, color='white')
    
    # Add annotations for metrics
    ax.text(0.5, -0.1, f"Precision: {precision:.2f}", ha='center', fontsize=12)
    ax.text(0.5, -0.15, f"Recall: {recall:.2f}", ha='center', fontsize=12)
    ax.text(0.5, -0.2, f"Accuracy: {accuracy:.2f}", ha='center', fontsize=12)
    
    # Formatting
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.3, 1.2)
    ax.set_aspect('equal', adjustable='box')
    ax.axis('off')
    ax.legend(loc="upper right")
    plt.title("Overlapping Squares for Precision, Recall, and Accuracy", fontsize=14)
    plt.show()

# Call the function with data
plot_overlapping_squares(TP, FP, FN, TN, total_samples)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the values for the confusion matrix
TP = 50  # True Positives
FP = 10  # False Positives
FN = 5   # False Negatives
TN = 100 # True Negatives

# Create a confusion matrix
confusion_matrix = np.array([[TP, FP],
                              [FN, TN]])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Create a grid of squares for the confusion matrix
for (i, j), value in np.ndenumerate(confusion_matrix):
    # Create a square patch for each element in the confusion matrix
    square = plt.Rectangle((j, i), 1, 1, fill=True, color='lightblue', edgecolor='black')
    ax.add_patch(square)
    # Add the value text in the center of the square
    ax.text(j + 0.5, i + 0.5, f'{value}', ha='center', va='center', fontsize=16)

# Set the limits and labels
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['Predicted Positive', 'Predicted Negative'])
ax.set_yticks([0.5, 1.5])
ax.set_yticklabels(['Actual Positive', 'Actual Negative'])

# Set title and show the plot
plt.title('Confusion Matrix Visualization using Square Patches')
plt.gca().invert_yaxis()  # Invert y-axis to have the actual positive on top
plt.grid(False)  # Disable the grid
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the values for the confusion matrix
TP = 50  # True Positives
FP = 10  # False Positives
FN = 5   # False Negatives
TN = 100 # True Negatives

# Calculate the total predictions
total_predictions = TP + FP + FN + TN

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Create a large square to represent total predictions
total_square = patches.Rectangle((0, 0), 1, 1, fill=True, color='lightgrey', alpha=0.5, label='Total Predictions')
ax.add_patch(total_square)

# Create squares for Actual Events (TP + FN) and Predicted Events (TP + FP)
actual_events_square = patches.Rectangle((0.1, 0.1), 0.8, 0.8, fill=True, color='blue', alpha=0.6, label='Actual Events (TP + FN)')
predicted_events_square = patches.Rectangle((0.1, 0.1), 0.8, 0.8, fill=True, color='red', alpha=0.4, label='Predicted Events (TP + FP)')

# Add squares to the plot
ax.add_patch(actual_events_square)
ax.add_patch(predicted_events_square)

# Add text annotations for each square
ax.text(0.5, 0.5, f'Total Predictions: {total_predictions}', ha='center', va='center', fontsize=12, color='black')
ax.text(0.5, 0.75, f'Actual Events: {TP + FN}', ha='center', va='center', fontsize=12, color='white')
ax.text(0.5, 0.25, f'Predicted Events: {TP + FP}', ha='center', va='center', fontsize=12, color='white')

# Set limits and labels
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Confusion Matrix Visualization using Overlapping Squares')

# Show the plot
plt.legend()
plt.show()
```
